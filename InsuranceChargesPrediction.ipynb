# Data Standardisation

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[['age','bmi','children']]=scaler.fit_transform(df[['age','bmi','children']])
```

# Training and Testing the model

## Training

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
x=df.drop('charges',axis=1)
y=df['charges']
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.drop('charges',axis=1),df['charges'],test_size=0.2,random_state=42)
```

```python
mlr=LinearRegression()
mlr.fit(x_train,y_train)
```

## Testing


```python
y_train_pred=mlr.predict(x_train)
```

```python
y_test_pred=mlr.predict(x_test)
```

# Model Evaluation

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import numpy as np
```

```python
mae=mean_absolute_error(y_train,y_train_pred)
mae1=mean_absolute_error(y_test,y_test_pred)
mse=mean_squared_error(y_train,y_train_pred)
mse1=mean_squared_error(y_test,y_test_pred)
rmse=np.sqrt(mse)
rmse1=np.sqrt(mse1)
r2=r2_score(y_train,y_train_pred)
r21=r2_score(y_test,y_test_pred)
```

```python
print(mae)
print(mae1)
print(mse)
print(mse1)
print(rmse)
print(rmse1)
print(r2)
print(r21)
```
